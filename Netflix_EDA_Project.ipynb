**Exploratory Data Analysis (EDA)on Raw Data**

**1. Movies vs TV Shows**

```python
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

**Observation:**

Movies are significantly higher in number compared to TV Shows on Netflix



**Insight:**

This indicates that Netflix focuses more on movie content to attract a wider audience.

**2.Top 10 Countries**

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar')
plt.title("Top 10 Countries with Most Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

**Observation:**

The United States has the highest amount of Netflix content

**Insight:**

This shows that the U.S. entertainment industry contributes heavily to Netflix’s library.

**3.Release Year Trend**

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(12,5))

plt.title("Netflix Content Release Trend")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

**Observation:**

Netflix content releases increased rapidly after 2015

**Insight:**

This suggests that Netflix expanded its global content production during recent years.

**4.Most Common Ratings**

```python
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)

plt.title("Most Common Ratings on Netflix")
plt.show()
```

**Observation:**

TV-MA is the most common rating on Netflix, showing that a large portion of content is designed for mature audiences.

**Insight:**

This suggests that Netflix mainly targets adult viewers by offering more mature and diverse entertainment content.

**Release Year Trend**

```python
plt.figure(figsize=(14,6))

sns.lineplot(
    x=df['release_year'].value_counts().sort_index().index,
    y=df['release_year'].value_counts().sort_index().values
)

plt.title("Netflix Content Release Trend", fontsize=16)
plt.xlabel("Release Year", fontsize=12)
plt.ylabel("Number of Shows", fontsize=12)

plt.grid(True)

plt.show()
```

**Observation:**

Netflix content releases increased rapidly after 2015, reaching a peak around recent years.

**Insight:**

This indicates that Netflix significantly expanded its content production and global streaming presence in recent years.

**Top 10 Geners On Netflix**

```python
plt.figure(figsize=(12,8))

top_genres = df['listed_in'].value_counts().head(10)

sns.barplot(
    x=top_genres.values,
    y=top_genres.index
)

plt.title("Top 10 Genres on Netflix", fontsize=16)
plt.xlabel("Number of Shows")
plt.ylabel("Genres")

plt.show()
```

**Observation:**

Drama, Comedy, and International Movies are among the most common genres on Netflix.

**Insight:**

**Pivot Table**

```python
content_year = df.pivot_table(
    index='release_year',
    columns='type',
    aggfunc='size',
    fill_value=0
)
```

**Heat Map**



```python
plt.figure(figsize=(12,8))

sns.heatmap(content_year, cmap='YlGnBu')

plt.title("Netflix Content Release Heatmap")

plt.show()
```

**Observation:**

The heatmap shows a significant increase in both Movies and TV Shows after 2015.

**Insight:**

This indicates that Netflix rapidly expanded its content library as global streaming demand increased.

**Netflix Dashboard Style Visualization**

```python
plt.style.use('dark_background')

fig, axes = plt.subplots(2, 2, figsize=(16,10))

# 1. Movies vs TV Shows
sns.countplot(
    x='type',
    data=df,
    ax=axes[0,0]
)

axes[0,0].set_title("Movies vs TV Shows")

# 2. Top Ratings
sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index[:5],
    ax=axes[0,1]
)

axes[0,1].set_title("Top Ratings")

# 3. Release Trend
release_trend = df['release_year'].value_counts().sort_index()

axes[1,0].plot(
    release_trend.index,
    release_trend.values
)

axes[1,0].set_title("Release Year Trend")

# 4. Top Genres
top_genres = df['listed_in'].value_counts().head(5)

axes[1,1].barh(
    top_genres.index,
    top_genres.values
)

axes[1,1].set_title("Top Genres")

plt.tight_layout()

plt.show()
```

**Observation:**

Movies dominate Netflix content, while Drama and Comedy remain highly popular genres. Content production also increased significantly after 2015

**Insight:**

The dashboard suggests that Netflix focuses on producing globally appealing entertainment content and continuously expands its streaming library to meet audience demand.
